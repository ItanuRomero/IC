```python
# With one music, we can filter by gender
blues_burning_df = results_df.loc[results_df['FirstImage'] == 'blues/Burning']
fig_3_df = blues_burning_df.sort_values(by=['Matches0.9'], ascending=False).head(10)
fig_3_df = fig_3_df[['SecondImage', 'NonZeroR', 'NonZeroG', 'NonZeroB', 'Matches0.8', 'Matches0.9', 'Matches1.0']]
fig_3_df = fig_3_df.rename(
    columns={'SecondImage': 'Música Comparada',
             'NonZeroR': 'AbsDiff R',
             'NonZeroG': 'AbsDiff G',
             'NonZeroB': 'AbsDiff B',
             'Matches0.8': 'ORB distância 0.8',
             'Matches0.9': 'ORB distância 0.9',
             'Matches1.0': 'ORB distância 1.0',
            }
)
fig_3_df = fig_3_df.set_index('Música Comparada')
dfi.export(fig_3_df, 'figura3.png')
fig_3_df
```

```python
fig_4_df = blues_burning_df.sort_values(by=['NonZeroG']).head(10)
fig_4_df = fig_4_df[['SecondImage', 'NonZeroR', 'NonZeroG', 'NonZeroB', 'Matches0.8', 'Matches0.9', 'Matches1.0']]
fig_4_df = fig_4_df.rename(
    columns={'SecondImage': 'Música Comparada',
             'NonZeroR': 'AbsDiff R',
             'NonZeroG': 'AbsDiff G',
             'NonZeroB': 'AbsDiff B',
             'Matches0.8': 'ORB distância 0.8',
             'Matches0.9': 'ORB distância 0.9',
             'Matches1.0': 'ORB distância 1.0',
            }
)
fig_4_df = fig_4_df.set_index('Música Comparada')
dfi.export(fig_4_df, 'figura4.png')
fig_4_df
```
