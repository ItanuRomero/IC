```python
sorted_by_good_matches = results[['Names', 'AreEqual', 'CountNonZeroAbsolute','GoodMatchesDistance0.5To1', 'Matches']]
sorted_by_good_matches = sorted_by_good_matches.drop(sorted_by_good_matches[sorted_by_good_matches.AreEqual == 'True'].index)
sorted_by_good_matches = sorted_by_good_matches.sort_values('CountNonZeroAbsolute', ascending=True)
by_absolute = list()
for index in sorted_by_good_matches.index:
    green = sorted_by_good_matches.iloc[index].loc['CountNonZeroAbsolute']
    green = green.replace('[', '').replace(']', '').split(',')[1]
    names = results.iloc[index].loc['Names'].split('/')
    names = [names[0], names[1].split('- ')[1]]
    by_absolute.append([index, names, green])
series = pd.Series(by_absolute)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(series)
```

```python
import ast
absolute = {
    'index': [],
    'red': [],
    'green': [],
    'blue': []
}
matches = {
    'index': [],
    '0.5': [],
    '0.6': [],
    '0.7': [],
    '0.8': [],
    '0.9': [],
    '1.0': []
}

def useAbsolute(rgb, index):
    list_rgb = ast.literal_eval(rgb)
    absolute['index'].append(index)
    absolute['red'].append(list_rgb[0])
    absolute['green'].append(list_rgb[1])
    absolute['blue'].append(list_rgb[2])
    

def useMatches(matches_list, index):
    list_matches = ast.literal_eval(matches_list)
    matches['index'].append(index)
    matches['0.5'].append(list_matches[0])
    matches['0.6'].append(list_matches[1])
    matches['0.7'].append(list_matches[2])
    matches['0.8'].append(list_matches[3])
    matches['0.9'].append(list_matches[4])
    matches['1.0'].append(list_matches[5])

    
for index in cleaned_df.index:
    record = cleaned_df.iloc[index]
    useAbsolute(record.loc['CountNonZeroAbsolute'], index)
    useMatches(record.loc['GoodMatchesDistance0.5To1'], index)
absolute_df = pd.DataFrame(absolute)
matches_df = pd.DataFrame(matches)
```
